## Joining Data with Geopandas

```python
#Geopandas joins are done the same as pandas
#First, I will read in data from cartographic boundary shapefiles
geometry = geopandas.read_file(r'C:\Users\GeogLabCommonAccess\Downloads\cb_2018_us_state_20m.shp')
geometry.head(1)
```

```python
#Next I will read in the next set of data that will be joined with the shapefile data read in in the previous code block
#This is election data previously used for an assignment
attribute_data = pd.read_csv('https://raw.githubusercontent.com/thomaspingel/geodata/master/election/state_election_data_1976-2016.csv')
attribute_data.head(1)
```

```python
#This is the election data merged to the state shapefile data
df = geometry.merge(attribute_data,how='left',left_on='STUSPS',right_on='state_po')
df.head(1)
```

# Creating plots in Geopandas

```python
# Geopandas can create plots because of its geospatial data library
#This is a simple example of a world plot
    
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))

_ = world.plot()
```

```python
#Accessing the world from the geopandas library then plotting only the USA and saving it
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
usa = world[world.name == "United States of America"]
_ = usa.simplify(tolerance=.25).plot()
plt.savefig('usa.png')
```

```python
#Plotting a buffer and saving it
_ = world.centroid.buffer(distance=4).plot(cmap="OrRd")
plt.savefig('buffermap.png')
```

# Doing a dissolve in pandas

```python
#Dissolving fields from a geopandas world plot
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
world = world[['continent', 'geometry']]
continents = world.dissolve(by='continent')
continents.plot()
continents.head()
```
